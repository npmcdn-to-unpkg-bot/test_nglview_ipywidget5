```python
import nglview as nv

view = nv.NGLWidget()
view
```

```python
view.add_component('../../../../ngl/data/1crn_apbs.pqr')
```

```python
view.add_component('../../../../ngl/data/1crn_apbs_pot.dxbin')
```

```python
view.parameters = dict(backgroundColor='black')
```

```python
view.clear_representations(component=0)
view.add_cartoon(component=0)
view.add_licorice(component=0, color_scheme='bfactor', side='double', radius=0.15)
```

```python
view.clear_representations(component=1)
view.add_surface(component=1, color='red', opacity=0.6, isolevel=-0.4, isolevel_type='value',
                 side='back')
view.add_surface(component=1, color='blue', opacity=0.6, isolevel=+0.4, isolevel_type='value',
                 side='front')
```
