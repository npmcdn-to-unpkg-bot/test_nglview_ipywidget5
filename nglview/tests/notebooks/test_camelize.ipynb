```python
import pytraj as pt
import nglview as nv
from IPython import display

traj = pt.datafiles.load_rna()

view = nv.show_pytraj(traj)
view
```

```python
view._clear_repr()
view.add_licorice()
```

```python
view.add_surface(opacity=0.3, colorScheme='residueindex')
```

```python
view.render_image()
```

```python
# make copy
image_data = view._image_data[:]
```

```python
display.Image(image_data)
```
