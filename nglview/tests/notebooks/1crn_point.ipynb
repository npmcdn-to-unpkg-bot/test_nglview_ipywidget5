```python
import nglview as nv

view = nv.NGLWidget()
```

```python
view.add_component('../../../../ngl/data/1crn.pdb')
view
```

```python
view.parameters = dict(backgroundColor='black')
```

```python
view.clear_representations()
view.add_point(point_size=7.0,
               opacity=0.5, 
               sort_particles=True,
               use_texture=True,
               clip_near=0.5,
               side='double',
               color_scheme='element',
               size_attenuation=True,
               edge_bleach=1.0,
               alpha_test=0.0)
view.add_licorice()
view.center_view()
```

```python
view.render_image()
```

```python
view._display_image()
```
