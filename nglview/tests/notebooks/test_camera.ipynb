```python
import nglview as nv
import nose.tools as nt
from traitlets import TraitError

view = nv.show_structure_file(nv.datafiles.PDB)
view
```

```python
view._get_full_params()
```

```python
msg = view._ngl_msg
assert msg['cameraType']['value'] == 'perspective'
```

```python
view.camera = 'orthographic'
```

```python
view._get_full_params()
```

```python
msg = view._ngl_msg
assert msg['cameraType']['value'] == 'orthographic'
```

```python
def func():
    view.camera = 'perspective_typos'
    
nt.assert_raises(TraitError, func)
```

```python
view2 = nv.show_structure_file(nv.datafiles.PDB)
# test setting camera before displaying it
view2.camera = 'orthographic'
view2
```

```python
view2._get_full_params()
```

```python
msg = view2._ngl_msg
assert msg['cameraType']['value'] == 'orthographic'
```

```python
# render
view.camera = 'orthographic'
view.render_image()
```
