```python
import pytraj as pt
import nglview as nv

traj = pt.load(nv.datafiles.TRR, nv.datafiles.PDB)

view = nv.show_pytraj(traj)
view
```

```python
view._clear_repr()
view.add_cartoon()
```

```python
view.render_image(factor=1, frame=3, transparent=False)
```

```python
# use ipywidgets.widget_image.Image
image = view.get_image()
image
```
