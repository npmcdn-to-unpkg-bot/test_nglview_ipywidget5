```python
import nglview as nv

from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
import os,time
print( time.asctime())

m2 = Chem.AddHs(Chem.MolFromSmiles('COc1ccc2[C@H](O)[C@@H](COc2c1)N3CCC(O)(CC3)c4ccc(F)cc4'))
m1 = Chem.AddHs(Chem.MolFromSmiles('Cc1ccccc1'))
AllChem.EmbedMultipleConfs(m1, useExpTorsionAnglePrefs=True, useBasicKnowledge=True)
AllChem.EmbedMultipleConfs(m2, useExpTorsionAnglePrefs=True, useBasicKnowledge=True)

view = nv.show_rdkit(m2, parmed=True)
view
```

```python
view.frame = 3
```

```python
view.orientation = [[-0.5069298386665028, 7.266124587163582, -13.331916845321544],
 [-0.0681890536168216, 0.8749214128697743, 0.479440063270481],
 [-0.015000343322753906, 0.11349999904632568, 0.13300001621246338],
 [0, 0, 0]]
```

```python
view.render_image()
```
